```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
with open('../backend/data/raw/Cambridge_Massachusetts_USA_network.pkl', 'rb') as f:
    G = pickle.load(f)

print(f"Network loaded: {len(G.nodes)} nodes, {len(G.edges)} edges")
```

```python
fig, ax = ox.plot_graph(G, node_size=0, edge_linewidth=0.5, figsize=(12, 12))
```

```python
import pandas as pd

edges_data = []
for u, v, data in G.edges(data=True):
    edges_data.append({
        'length': data.get('length', 0),
        'highway': data.get('highway', 'unknown'),
        'maxspeed': data.get('maxspeed', 'unknown'),
        'lanes': data.get('lanes', 'unknown')
    })

df = pd.DataFrame(edges_data)
print(df.describe())
print(df['highway'].value_counts())
```
